```python
def info_collection_in_dataframe(url):
    response=requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    data_info={
        # "customer_name":[],
        "date_of_review":[],
        "review_title":[],
        "review_text":[],
    }

    #User Details 
    user_details=soup.find_all("h3", class_="text_sub_header userStatusWrapper")
    for user_detail in user_details:
        username=user_detail.span.text
        review_date=user_detail.time.text
        print(username)
        # data_info["customer_name"]=username
        data_info["date_of_review"].append(review_date)

    ## Review title 
    review_titles=soup.find_all("h2", class_="text_header")
    for review_title in review_titles:
        review_title=review_title.text
        data_info["review_title"].append(review_title)

    ## sub Header user 
    sub_headers=soup.find_all("h3", class_="text_sub_header userStatusWrapper")
    review_texts=soup.find_all("div", class_="text_content")
    for review in review_texts:
        review=review.text
        data_info["review_text"].append(review)
    return data_info
    
    
    
```

```python
pd.DataFrame(info_collection_in_dataframe("https://www.airlinequality.com/airline-reviews/british-airways/page/1"))
```
